你是中山大学的一名金融学教授，现在正在设计「金融数据分析与建模」课程的教学内容。请根据学生的需求和教学进度的安排，做一个教学大纲，包含课程目标、主要章节和每章的关键知识点。

## 学生

- 金融专业硕士。
  - 约有 70% 的学生来自金融和经济专业，30% 来自计算机科学、统计学和其他专业。
  - 多数学生本科毕业于 985 或 211 大学，具有较好的自学能力。
  - 这些学生前期修过概率统计、线性代数、微积分等基础课程；也修过金融学原理、投资学、公司理财等专业课程；约有半数的学生修过计量经济学课程 (覆盖了线性回归模型、假设检验、静态面板数据模型等内容)，但对于 Monte Carlo、Bootstrap、DID、Matching，RCM 等方法了解有限。多数人都没有修过时间序列方面的内容。数据处理的经验也非常有限。
  - 具备一定的编程基础，约有 60% 的学生会用 Python，80% 的同学熟悉 Stata 的常用命令。

## 课程目标
- 掌握金融数据分析与建模的基本方法和工具。
- 熟悉 Python 在金融数据处理、分析和建模中的应用。
- 能够独立完成金融数据分析项目，从数据获取、清洗、分析到建模和结果解释。
- 理解金融市场的基本原理和机制，能够将理论知识应用于实际问题的解决。
- 课程结束时，学生应能够胜任金融数据分析相关的实习或工作岗位。比如，能独立完成如下场景的数据分析和建模任务：
  - 投资银行：进行市场分析、风险评估和投资组合优化。
  - 资产管理公司：分析基金表现、优化资产配置和评估投资策略
  - 风险管理部门：识别和评估金融风险，制定风险控制策略。
  - 政府金融机构和研究部门：进行经济数据分析，采用统计和可视化方法搞清楚数据事实，能够用计量经济学方法分析政策效果。
  - 金融科技公司：开发和应用金融数据分析工具，进行大数据分析和机器学习建模。
  - 文本数据分析：处理和分析金融新闻、社交媒体数据，提取有价值的信息用于投资决策。

## 课程素材

- 数据：多数案例需采用中国市场数据，部分案例可使用国际市场数据。
  - CSMAR 数据库（需申请使用权限）：这些学生都有 CSMAR 数据库的使用权限，可以利用该数据库获取中国股票市场、债券市场、基金市场等多种金融数据。但需要搞清楚如何使用 CSMAR 数据库的 API 进行数据获取。
  - Wind 数据库（需申请使用权限）：部分学生也有 Wind 数据库的使用权限，可以利用 Wind 数据库获取更全面的金融数据。
  - AKshare 数据库（免费）：可以使用 AKshare 库获取部分中国市场的金融数据。
  - Yahoo Finance 数据（免费）：可以使用 yfinance 库获取国际市场的金融数据。目前来看，在中国大陆，使用 Python 获取 Yahoo Finance 数据很不稳定，经常难以获取较长的数据。
  - TuShare 数据（免费/付费）：可以使用 TuShare 库获取中国市场的金融数据。免费版数据有限，建议申请付费版以获取更全面的数据。
- 工具：Python 作为主要编程语言，使用 Jupyter Notebook 进行代码。

## 授课安排

- 共 12 次课，分 6 周完成。
  - 共 36 学时，每周上 2 次课，每次 3 节课*45 分钟，4.5 小时，共 6 周
- 授课安排和作业：讲 10 个 Topics，留出 2 次课进行小组展示和课程总结。
  - 前 6 次课：全程由教师授课 3 节课。
    - 第 1-5 次课，每次课后布置个人作业，要求学生在一周内完成。
    - 第 6 次课，布置小组作业，要求学生在接下来的一周内完成小组展示内容的准备。
  - 第 7-12 次课：老师讲授 2 节课，学生展示 1 节课。
    - 展示内容为上一周布置的小组作业，每次挑选 2-3 个小组进行展示，其他学生和老师进行点评和讨论。
  - 课程结束 2 周内，学生提交课程项目报告。
- 成绩构成：
  - 个人作业：30%
  - 小组展示：30%
  - 课程项目报告：40%

### 参考资料：
- 2-3 本相关的教材或参考书目：要能覆盖金融数据分析、Python 编程和金融建模等内容；其中有一本可以完全是理论性的教材，另外 1-2 本要包含大量 Python 代码示例和实操内容。
- 相关的在线资源或文档链接，最好是 github 上的开源项目或教程。
  - 最好是使用 .ipynb 格式的 notebook 文件作为参考资料。
- 推荐的数据集来源或 API 接口，方便学生获取实际的金融数据进行练习和项目开发。
