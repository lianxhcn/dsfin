## 任务

python 代码

把当前文件夹下所有 '.svg' 文件转换为 '.png' 文件，转换后的文件保存在同一目录下。

1. 原始 .svg 文件不要删除
2. 转换后的文件加前缀 'data_manage_'，最终 {fig_name} = "'data_manage_' + {原文件名}" (不用加后缀 .png)
3. 生成 .png 图片文件的 Markdown 格式的链接文本，格式如下：
```markdown
![{设定一个合适的中文解释性文字，20字以内}](https://fig-lianxh.oss-cn-shenzhen.aliyuncs.com/{fig_name}.png){width="90%"}

![{设定一个合适的英文解释性文字，20字以内}](https://fig-lianxh.oss-cn-shenzhen.aliyuncs.com/{fig_name}.png){width="90%"}
```

````python
import os
import subprocess

# ========== 配置 ==========
# Inkscape 路径（根据实际安装位置修改）
INKSCAPE = r"C:\Program Files\Inkscape\bin\inkscape.exe"

# 如果不确定路径，可以先测试：
if not os.path.exists(INKSCAPE):
    # 尝试其他常见路径
    for path in [
        r"C:\Program Files\Inkscape\bin\inkscape.exe",
        r"C:\Program Files (x86)\Inkscape\bin\inkscape.exe",
        r"C:\Program Files\Inkscape\inkscape.exe",
    ]:
        if os.path.exists(path):
            INKSCAPE = path
            break
    else:
        # 如果都找不到，尝试直接调用（要求 Inkscape 在 PATH 中）
        INKSCAPE = "inkscape"

print(f"使用 Inkscape: {INKSCAPE}\n")

# ========== 主程序 ==========
svg_files = [f for f in os.listdir('.') if f.lower().endswith('.svg')]

if not svg_files:
    print("当前目录下没有找到 .svg 文件。")
else:
    print(f"找到 {len(svg_files)} 个 SVG 文件，开始转换...\n")
    markdown_links = []

    for svg_file in sorted(svg_files):
        base_name = os.path.splitext(svg_file)[0]
        fig_name = f"data_manage_{base_name}"
        png_file = f"{fig_name}.png"

        abs_svg = os.path.abspath(svg_file)
        abs_png = os.path.abspath(png_file)

        try:
            result = subprocess.run(
                [
                    INKSCAPE,
                    abs_svg,
                    "--export-type=png",
                    f"--export-filename={abs_png}",
                    "--export-dpi=150"
                ],
                capture_output=True,
                text=True,
                timeout=60
            )

            if os.path.exists(png_file):
                print(f"✅ {svg_file} -> {png_file}")

                url = f"https://fig-lianxh.oss-cn-shenzhen.aliyuncs.com/{fig_name}.png"
                desc_cn = base_name.replace('_', ' ').replace('-', ' ').strip()

                md_cn = f'![{desc_cn} 示意图]({url}){{width="90%"}}'
                md_en = f'![{desc_cn} illustration]({url}){{width="90%"}}'
                markdown_links.append((svg_file, md_cn, md_en))
            else:
                print(f"❌ {svg_file}: 转换失败")
                if result.stderr:
                    print(f"   错误: {result.stderr[:200]}")

        except Exception as e:
            print(f"❌ {svg_file}: {e}")

    # 输出 Markdown 链接
    print("\n" + "=" * 70)
    print("Markdown 链接文本（可直接复制）：")
    print("=" * 70 + "\n")

    for svg_file, md_cn, md_en in markdown_links:
        print(f"<!-- 源文件: {svg_file} -->")
        print(md_cn)
        print(md_en)
        print()

    print(f"共转换 {len(markdown_links)}/{len(svg_files)} 个文件。")
````
