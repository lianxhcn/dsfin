## 检查数据

```python
import pandas as pd
import glob
from pathlib import Path

base = Path("data_raw")

print("=" * 55)

# 个股日度
files = sorted(glob.glob(str(base / "stock_daily/*.csv")))
print(f"stock_daily：{len(files)} 个文件")
if files:
    dfs = [pd.read_csv(f) for f in files]
    df_all = pd.concat(dfs, ignore_index=True)
    date_col = [c for c in df_all.columns if 'date' in c.lower() or '日期' in c][0]
    print(f"  总行数：{len(df_all):,}")
    print(f"  时间范围：{df_all[date_col].min()} ~ {df_all[date_col].max()}")
    print(f"  列名：{list(df_all.columns)}")

print()

# 指数
files2 = sorted(glob.glob(str(base / "index_daily/*.csv")))
print(f"index_daily：{len(files2)} 个文件")
if files2:
    df2 = pd.read_csv(files2[0])
    print(f"  列名：{list(df2.columns)}")

print()

# 宏观
files3 = sorted(glob.glob(str(base / "macro_monthly/*.csv")))
print(f"macro_monthly：{len(files3)} 个文件")
for f in files3:
    df3 = pd.read_csv(f)
    print(f"  {Path(f).name}：{len(df3)} 行，列名：{list(df3.columns)}")

print()

# 财务年度
p_fin = base / "fin_annual/fin_indicators.csv"
if p_fin.exists():
    df4 = pd.read_csv(p_fin)
    print(f"fin_annual：{len(df4)} 行")
    print(f"  列名：{list(df4.columns)}")
    print(f"  年份：{sorted(df4['year'].unique()) if 'year' in df4.columns else '未知'}")
else:
    print(f"fin_annual：文件不存在，检查路径")
    print(f"  实际存在的文件：{list((base/'fin_annual').glob('*')) if (base/'fin_annual').exists() else '目录不存在'}")

print()

# 公司信息
p_co = base / "company_info.csv"
if p_co.exists():
    df5 = pd.read_csv(p_co)
    print(f"company_info：{len(df5)} 行")
    print(f"  列名：{list(df5.columns)}")
    print(df5.to_string(index=False))
else:
    print("company_info.csv：不存在")
    print(f"  data_raw 下的文件：{[f.name for f in base.glob('*')]}")
```
