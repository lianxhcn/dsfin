## 讲义：金融计量中的Bootstrap方法

### 第一部分：Bootstrap的基本思想——从一个简单例子开始

首先，我们通过一个简单的例子来理解Bootstrap的核心思想。

**1. 什么是总体和样本？**
在统计学中，我们通常会区分**总体（Population）**和**样本（Sample）**。
*   **总体**指的是我们感兴趣的所有个体或观测值的集合。例如，中国所有上市公司的市盈率。
*   **样本**是从总体中抽取的一个子集，我们通过分析样本的特征来推断总体的特征。例如，从上交所随机抽取50家上市公司的市盈率数据。

**2. 传统抽样与Bootstrap抽样**
假设我们已经从总体中抽取了一个原始样本。现在，我们想了解样本中某个统计量（如样本均值、中位数或回归系数）的**抽样分布（Sampling Distribution）**，即如果重复从总体中抽取样本，这个统计量会如何波动。

*   **传统抽样（不放回）**：传统的抽样方式是从总体中抽取个体，并且一旦被抽取，该个体就不会被再次抽取（即**不放回抽样**）。例如，从6个观测值中抽取5个，每个值只出现一次。这种抽样常用于“抽签”等场景。
*   **Bootstrap抽样（有放回重抽样）**：与传统抽样不同，Bootstrap的核心操作是**从我们已有的原始样本中，进行有放回的随机抽样**，抽取与原始样本大小相同的新样本。这些新样本被称为**Bootstrap样本**。

**3. 简单例子**
让我们来看一个具体的例子。假设我们有一个包含6个观测值的原始样本 `x`:
`x = (1, 2, 3, 4, 5, 9)`

如果我们进行一次**传统抽样**（假设抽取5个观测值）:
一个可能的样本是：`(1, 2, 3, 4, 5)`

现在，我们进行一次**Bootstrap抽样**（从原始样本 `x` 中有放回地抽取6个观测值）:
一个可能的Bootstrap样本是：`(1, 9, 4, 3, 4, 3)`
请注意，在这个Bootstrap样本中：
*   观测值 `3` 和 `4` 各出现了两次。
*   观测值 `2` 和 `5` 没有出现。
*   观测值 `1` 和 `9` 各出现了一次。
这种“有放回”的特性是Bootstrap区别于传统抽样的关键。

**4. Bootstrap的核心思想**
Bootstrap方法的核心思想是：**由于我们无法直接从未知的总体中进行重复抽样以了解统计量的抽样分布，那么我们可以退而求其次，通过从我们已知的原始样本（将其视为一个“经验总体”或“经验分布函数，EDF”）中进行有放回的重复抽样，来近似模拟统计量的抽样分布**。EDF将原始样本中的每个观测值赋予 $n^{-1}$ 的等概率。通过分析这些数千次（甚至更多）生成的Bootstrap样本上的统计量分布，我们可以估计原始统计量的标准误差、偏差，构建置信区间，或进行假设检验。这是一种**“用计算代替理论推导”**的强大策略。

### 第二部分：统计推断的两种基本方法

在金融计量中，我们进行统计推断（如参数估计、假设检验、置信区间构建）主要依赖于两种基本方法：

**1. 基于渐近分布的传统推断方法**
*   **基本原理**：这类方法依赖于**大样本理论**（如中心极限定理、大数定律）来近似统计量的抽样分布。当样本量足够大时，许多统计量（尤其是经过标准化后的**枢轴量**，即其分布不依赖于任何未知参数的量）的分布会收敛到某些**已知且简单的理论分布**（如正态分布 $N(\mu, \sigma^2)$、卡方分布 $\chi^2(k)$、F分布或t分布 $t(k)$）。我们利用这些已知分布的临界值或概率来构建置信区间和进行假设检验。
*   **局限性**：
    *   **有限样本表现不佳**：在样本量较小，或数据分布严重偏离正态性、同方差性、无自相关等假设时，渐近近似可能不准确。这可能导致置信区间的实际覆盖率与名义覆盖率不符，或假设检验的I类错误率（犯弃真错误）失控。
    *   **复杂统计量推导困难**：对于非线性模型、非标准统计量（如某个复杂比率、中位数）或某些复杂的估计方法（如广义矩估计量GMM），推导其精确的有限样本分布或渐近分布可能非常复杂，甚至无法得到解析解。

**2. 基于重抽样（Re-sampling）的推断方法——Bootstrap属于此类**
*   **基本原理**：这类方法通过从原始数据中反复抽取（通常是有放回地）新的样本，来**直接模拟统计量的抽样分布**，从而避免或减少对特定理论分布形式的依赖。
*   **Bootstrap的定位**：Bootstrap方法是这类重抽样方法中最流行和广泛应用的一种。它通过**计算代替复杂的数学分析**，极大地拓展了统计推断的应用范围。
*   **优势**：
    *   **更准确的有限样本推断**：相比渐近方法，Bootstrap在小样本或非标准分布情况下通常能提供**更准确的分布近似**、置信区间覆盖率和假设检验的I类错误率。
    *   **适用性广**：它能够应用于各种复杂的统计量（如回归系数、R方、F统计量、中位数等），无需为每个统计量推导复杂的解析公式。
    *   **对模型假设依赖少**：尤其是非参数Bootstrap，它直接利用数据的经验分布，对总体分布的形式假设较少，因此**对模型误设具有更强的鲁棒性**。

**3. 标准差（SD）与标准误（SE）的区别与联系，以及Bootstrap在此的作用**
在统计推断中，我们经常会遇到“标准差（Standard Deviation, SD）”和“标准误（Standard Error, SE）”这两个概念。

*   **标准差（SD）**：**描述个体数据值**与均值的离散程度。例如，如果一个班级的考试成绩平均分为80分，标准差为10分，这意味着大多数学生的成绩在70-90分之间波动。SD衡量的是样本内部个体观测值的变异性。
*   **标准误（SE）**：描述**样本统计量**（如样本均值、回归系数）的抽样分布的离散程度，即样本统计量与总体真实参数的接近程度。它是统计量估计值的标准差。例如，如果我们在不同批次抽取多个样本，并计算每个样本的均值，这些样本均值会有一个自身的分布，其标准差就是样本均值的标准误。SE衡量的是统计量在重复抽样下的稳定性。
    *   两者之间的关系：对于样本均值，理论上`SE = SD/sqrt(N)`。

*   **Bootstrap在此的作用**：
    传统上，SE的计算通常需要对总体分布的参数进行估计，或者依赖渐近理论的公式。当这些假设难以满足或统计量复杂时，估算SE变得困难。
    **Bootstrap提供了一种不依赖这些假设，通过计算直接估计SE的通用方法**。其步骤是：
    1.  从原始样本中生成 $J$ 个（例如 $J=5000$ 或 $10000$ 个）Bootstrap样本。
    2.  对每个Bootstrap样本，计算我们感兴趣的统计量 $Q$，得到一系列Bootstrap统计量：$Q_1, Q_2, ..., Q_J$。
    3.  这些Bootstrap统计量 $Q_j$ 的标准差 $SD(Q_1, Q_2, ..., Q_J)$，就是统计量 $Q$ 的**Bootstrap标准误**。
    这种方法被称为**Bootstrap标准误差**，是传统渐近标准误差的一种替代方法。

### 第三部分：Bootstrap的核心假设与适用/不适用场景

Bootstrap作为一种强大的统计推断工具，并非万能。它有其核心假设，并在此基础上表现出优越性或局限性。理解这些假设有助于我们正确选择和应用Bootstrap方法，并对其结果进行合理评估。

**1. Bootstrap的核心假设**

Bootstrap方法的核心在于使用**经验分布函数（EDF）**或**参数模型**来模拟未知的真实总体分布。其有效性主要依赖于以下几个核心假设：

*   **原始样本具有代表性 (Representativeness of Original Sample)**：
    *   这是最基本且最重要的假设。Bootstrap通过从原始样本中进行有放回的重抽样来生成Bootstrap样本，从而近似总体的抽样过程。因此，**原始样本必须是从总体中随机抽取并能很好地代表总体分布**。如果原始样本本身存在严重的偏倚，那么Bootstrap结果也将是偏倚的。
    *   在非参数Bootstrap中，这一假设通常意味着观测值是**独立同分布（Independent and Identically Distributed, IID）**的。这意味着每个观测值都是相互独立的，并且来自相同的未知总体分布。如果数据并非IID（例如存在异方差、序列相关、分组相关等），则需要对传统的非参数Bootstrap进行修正，否则朴素的Bootstrap应用可能会失败。

*   **统计量的渐近分布性质**：
    *   虽然Bootstrap方法旨在避免对统计量渐近分布形式的强假设，但其理论有效性（尤其是“渐近精确性”或“渐近改进”）往往是在大样本下建立的。
    *   对于许多常见的统计量（如样本均值、回归系数），当样本量足够大时，其标准化形式（即“中心化”和“学生化”后的统计量）会趋向于一个**渐近枢轴量（Asymptotically Pivotal Statistic）**。这意味着该统计量的渐近分布不再依赖于任何未知参数。如果统计量是渐近枢轴的，Bootstrap能够更准确地近似其有限样本分布，从而提供渐近修正。

*   **总体分布的平稳性 (Stationarity)**：
    *   在处理时间序列数据时，如果采用传统的非参数Bootstrap（不考虑时间序列结构），则隐含地假设时间序列是平稳的。更一般的块Bootstrap方法，则试图在保持时间依赖性的前提下进行重抽样，通常假设时间序列具有某种平稳性特征，例如弱平稳性。

**2. Bootstrap的适用场景**

Bootstrap方法因其灵活性和鲁棒性，在以下多种场景中非常适用：

*   **标准统计推断方法失效或难以应用时**：
    *   **小样本情形**：当样本量较小，大样本渐近理论提供的近似可能不准确时。
    *   **复杂统计量**：对于均值、中位数、分位数、比率、回归系数等复杂统计量的抽样分布，难以通过解析方法推导时。例如，中位数的方差估计 或某些非线性函数估计量的标准误。
    *   **非正态或未知分布**：数据不服从正态分布，或总体分布形式未知时。
    *   **非标准模型**：在非线性模型、半参数模型或非参数模型中进行推断时。

*   **需要稳健估计标准误、构建置信区间或进行假设检验时**：
    *   **异方差和序列相关**：对于存在异方差或序列相关的回归模型，传统的标准误估计可能失效。通过Pairs Bootstrap、Wild Bootstrap或Block Bootstrap等方法可以得到对这些问题鲁棒的推断。
    *   **模型假设较少**：特别是非参数Bootstrap，直接利用数据的经验分布，对总体分布的形式假设最少，因此**对模型误设具有更强的鲁棒性**。

*   **作为诊断工具**：Bootstrap可以用于评估统计推断对个体观测值的敏感性。例如，通过“jackknife-after-bootstrap”方法来检测异常值的影响。

**3. Bootstrap的局限性与不适用场景**

尽管Bootstrap用途广泛，但它并非没有局限性，在某些情况下应用不当可能导致错误结果：

*   **原始样本不具代表性**：如果原始样本不能很好地代表总体，那么Bootstrap样本也将无法反映真实的总体分布，从而导致推断失效。
*   **重尾分布与无限方差**：如果数据来自具有无限方差的重尾分布，朴素的Bootstrap可能会不一致。
*   **极端值问题**：当原始样本中存在极值或离群点时，由于Bootstrap是有放回抽样，这些极值在Bootstrap样本中出现的频率可能会被放大或缩小，从而影响估计的准确性。
*   **依赖性结构处理不当**：
    *   对于时间序列数据，如果存在时间依赖性（如自相关），而使用简单的IID非参数Bootstrap，会破坏这种依赖结构，导致结果不准确。需要采用专门为时间序列设计的Bootstrap方法（如Block Bootstrap、Sieve Bootstrap或Autoregressive Bootstrap）。
    *   对于面板数据，如果存在组内相关性（如个体固定效应），同样需要采用专门的方法（如Cluster Bootstrap）来保留这种依赖性 。
*   **参数在参数空间边界上**：当参数的真实值位于参数空间的边界时，Bootstrap可能不一致。
*   **非渐近枢轴统计量**：对于非渐近枢轴的统计量，朴素的Bootstrap可能无法提供渐近修正。虽然双重Bootstrap (Double Bootstrap) 等迭代方法可以尝试创建渐近枢轴量，但其计算成本更高。
*   **计算成本高昂**：对于一些复杂的模型或需要进行多层重抽样（如双重Bootstrap）的场景，Bootstrap的计算成本会非常高，可能需要数小时甚至更长时间。

### 第四部分：常用的Bootstrap方法

Bootstrap方法家族庞大，针对不同的数据结构和推断目标，发展出了多种变体。以下介绍几种在金融计量中常用的Bootstrap方法：

**1. 非参数Bootstrap（Nonparametric Bootstrap）**
*   **原理**：这是最基本、最常用的Bootstrap类型。它直接从原始样本的**经验分布函数（Empirical Distribution Function, EDF）**中进行**有放回的随机抽样**。每个原始观测值被赋予 $n^{-1}$ 的概率，从中抽取与原样本大小相同的 $n$ 个观测值，构成一个Bootstrap样本。
*   **应用**：由于其通用性，非参数Bootstrap适用于大多数情况，例如估计参数的方差、偏差，构建置信区间，以及进行假设检验。

**2. 参数Bootstrap（Parametric Bootstrap）**
*   **原理**：与非参数Bootstrap不同，参数Bootstrap首先根据原始数据拟合一个**参数模型**（例如，假设误差项服从正态分布，并根据数据估计其均值和方差）。然后，它从这个**已拟合的参数模型**中生成Bootstrap样本，而非直接从原始数据中抽样。
*   **适用场景**：当研究者对数据背后的总体分布形式有较强的理论假设时（如回归模型中的误差项服从正态分布），使用参数Bootstrap可以提高效率。例如，航空空调数据在指数模型下的参数模拟。

**3. 残差重抽样Bootstrap（Residual Resampling Bootstrap）**
*   **原理**：主要应用于回归模型。首先，研究者使用原始数据估计回归模型并得到模型的残差（或经过修正的残差）。然后，从这些残差中进行**有放回抽样**以生成新的误差项，再结合原始的自变量值和估计的回归系数来生成新的因变量，从而构建Bootstrap样本。
*   **适用场景**：当回归模型的**自变量被视为固定设计**，且误差项满足**独立同分布（Independent and Identically Distributed, IID）**的假设时（特别是同方差假设）。
*   **局限性**：当回归模型存在**异方差**或**序列相关**时，标准的残差重抽样可能不适用，因为它隐含地假定误差是IID的。

**4. 成对重抽样Bootstrap（Pairs Bootstrap / Case Resampling）**
*   **原理**：在回归模型中，成对重抽样不是单独抽样残差，而是直接从原始数据中的**观测对**（即每个(因变量Yi, 自变量Xi)对）中进行有放回抽样。每次抽取一个完整的观测对来构成Bootstrap样本。
*   **优势**：相比残差重抽样，成对重抽样对**异方差**和误差分布的假设更少，因为它保留了原始数据中因变量和自变量之间的任何潜在依赖关系。因此，它能够更好地处理异方差问题。
*   **适用场景**：当**自变量是随机的**，或存在异方差，或对误差分布的形式知之甚少时。

**5. 块重抽样Bootstrap（Block Bootstrap）**
*   **原理**：专门为**时间序列数据或存在序列相关的数据**设计。它将原始时间序列数据分割成连续的“块”（Block），然后对这些块进行有放回抽样，再将抽取的块按照顺序拼接起来形成新的Bootstrap样本。
*   **目的**：通过抽取数据块而非单个观测值，来**捕获和保留数据中的时间依赖性**或序列相关结构。
*   **类型**：
    *   **移动块Bootstrap（Moving Block Bootstrap，MBB）**：块之间存在重叠。
    *   **平稳Bootstrap（Stationary Bootstrap）**：块的长度是随机的，以确保生成的Bootstrap序列是（渐近）平稳的，并减少块边界效应。

**6. 野性Bootstrap（Wild Bootstrap）**
*   **原理**：在回归模型存在**异方差**但其具体形式未知时，野性Bootstrap是一种有效的修正方法。它不对残差进行直接重抽样，而是对每个残差 $\hat{e}_i$ 乘以一个均值为0、方差为1的随机变量 $\epsilon_i^*$ （例如，Rademacher分布，其值为1或-1各以0.5的概率取值；或标准正态分布），以生成新的误差项 $\hat{e}_i^* = \hat{e}_i \cdot \epsilon_i^*$。
*   **目的**：这种方法可以在不对异方差的具体形式进行建模的情况下，**保留原始数据中的异方差结构**，从而提供对异方差鲁棒的推断。

**7. 迭代Bootstrap（Iterated Bootstrap）/ 双重Bootstrap（Double Bootstrap）/ 预枢轴（Prepivoting）**
*   **原理**：这类方法涉及两个层次的重抽样。首先进行第一层Bootstrap抽样，得到一组Bootstrap样本。然后，**对每个第一层得到的Bootstrap样本，再进行第二层Bootstrap抽样**。
*   **目的**：
    *   **修正偏差**：用于进一步修正Bootstrap估计量本身的偏差。
    *   **改进置信区间**：显著提高置信区间的覆盖率准确性，使其更接近名义水平。
    *   **评估敏感性**：用于诊断和评估Bootstrap计算对个体观测值的敏感性。
*   **代价**：由于涉及两层循环，其计算成本非常高，可能需要大量的模拟次数 (例如，R次第一层模拟，每次M次第二层模拟，总计R×M次模拟)。

**8. 平衡Bootstrap（Balanced Bootstrap）**
*   **原理**：这是一种提高蒙特卡洛模拟效率的技术。在进行Bootstrap抽样时，平衡Bootstrap会确保**原始样本中的每个观测值在所有Bootstrap样本中出现的总频率是相等的**（例如，如果进行了R次Bootstrap模拟，则每个原始观测值总共被抽取R次）。
*   **目的**：通过“平衡”抽样过程，可以显著减少蒙特卡洛模拟误差，尤其在估计统计量的偏差或方差时，可以**显著提高估计效率**。

### 第五部分：Bootstrap的常见问题与实践考量

在实际应用Bootstrap方法时，同学们常常会遇到一些具体问题。下面，我们针对几个常见问题进行解答和讨论。

**1. Bootstrap的次数（$B$）多少为宜？**

Bootstrap的模拟次数（通常用 $B$ 或 $R$ 表示）是一个重要的实践问题。一般来说，$B$ 的选择需要权衡计算成本和结果的精确性。

*   **估计标准误（Standard Error）**：
    *   如果目标仅仅是估计统计量的标准误，通常较少的 $B$ 值即可接受。例如，Efron（1993）建议**50到200次重抽样**可能就足够了。
    *   在Stata中进行OLS回归时，默认的Bootstrap次数通常为50次。
*   **构建置信区间（Confidence Intervals）**：
    *   对于更精确的置信区间估计，特别是涉及百分位数方法时，需要更多的Bootstrap样本。建议通常为**5000到10000次**。
    *   更高的次数可以更好地捕捉统计量分布的尾部特征，从而提供更准确的置信区间覆盖率。
*   **估计偏差（Bias）或描述完整分布**：
    *   如果需要精确估计统计量的偏差，或者要详细描述统计量的完整抽样分布（例如，通过直方图或密度图），建议进行**10000次或更多**的重抽样。
*   **假设检验（Hypothesis Testing）中的p值**：
    *   对于假设检验中p值的计算，特别是当p值非常小时，需要更多的Bootstrap次数来确保其准确性。通常也建议**5000次以上**。
    *   有研究指出，在实际应用中，**1000次**的Bootstrap样本量已经足够用于许多场景。考虑到现代计算机的计算能力，使用 $B=1000$ 或 $B=2000$ 已经非常普遍，且不会带来过高的计算时间成本。
*   **计算效率与精度**：
    *   如果统计量计算成本很高（例如，涉及复杂优化），或者进行多层Bootstrap（如双重Bootstrap），可能需要采取一些**提高计算效率的方法**，例如平衡Bootstrap（Balanced Bootstrap） 或重要性重抽样（Importance Resampling），以在给定模拟次数下提高精度，或者减少所需模拟次数。

**2. 有时候我通过设定特定的种子值（Seed Value）可以得到更好的结果，这样是否合理？**

设定特定的种子值（`set seed` 命令）是完全合理且推荐的做法，但其目的并非是为了“得到更好的结果”，而是为了**确保结果的可重现性（Reproducibility）**。

*   **可重现性**：Bootstrap方法依赖于伪随机数生成器。每次运行如果没有设置种子值，都会生成不同的随机序列，从而导致Bootstrap样本和最终的统计量估计（如标准误、置信区间）略有不同。设定种子值可以固定随机数生成器的起始点，使得每次运行都能得到**完全相同**的结果。这对于学术研究的验证和同行评审至关重要。
*   **并非“更好”**：种子值的设定本身并不能使Bootstrap结果在统计意义上变得“更好”或更准确。如果通过特定种子值偶然获得了“看起来更好”的结果，这可能只是随机波动的结果，或者暗示了统计量对随机抽样的高度敏感性。在这种情况下，正确的做法应该是：
    1.  **增加Bootstrap的次数 $B$**：通过更多的重复，减少蒙特卡洛模拟误差，使结果更稳定，减少对特定随机抽样路径的依赖。
    2.  **检查统计量的稳定性**：如果统计量在不同随机种子下表现出很大差异，这可能表明原始样本量不足，或统计量本身对数据中的微小扰动非常敏感。
    3.  **使用更稳健的Bootstrap方法**：例如，如果数据存在异方差，从参数模型中重抽样可能比非参数方法更有效。

**3. 对于时间序列和面板数据，如何进行Bootstrap比较好？**

时间序列数据和面板数据都包含重要的依赖结构，直接使用传统的IID非参数Bootstrap会破坏这些结构，导致推断失效。因此，需要采用专门设计的Bootstrap方法。

*   **时间序列数据（Time Series Data）**：
    *   **问题**：时间序列数据存在时间相关性（如自相关），即观测值之间不独立。
    *   **解决方案**：
        1.  **块Bootstrap（Block Bootstrap）**：这是最常用的方法之一。它通过重抽样原始时间序列中的连续数据块（Block）来保留数据中的时间依赖性。
            *   **移动块Bootstrap（Moving Block Bootstrap，MBB）**：块之间存在重叠。这种方法通常表现更好，因为它能生成更多的独特块，更好地捕捉依赖性。
            *   **平稳Bootstrap（Stationary Bootstrap）**：这是MBB的一种变体，块的长度是随机的（通常服从几何分布），以确保生成的Bootstrap序列是（渐近）平稳的，并减少块边界效应。
            *   **块长度的选择**：块的长度是一个关键参数。过短的块无法捕捉长期依赖性，过长的块会导致样本量不足。选择最优块长度是一个活跃的研究领域，但没有普遍适用的简单规则。
        2.  **Sieve Bootstrap（筛法Bootstrap）**：这种方法通过拟合一个参数化的时间序列模型（如自回归模型AR(p)）来近似误差项的依赖结构，然后从拟合模型的残差中重抽样，并用这些重抽样的残差来生成新的序列。这种方法在某些情况下被认为优于块Bootstrap，因为它能更好地捕捉自相关结构。
        3.  **自回归（Recursive/Autoregressive）Bootstrap**：在动态回归模型中，可以根据估计的回归模型和残差生成Bootstrap数据。如果误差项服从已知的参数分布，则可以从该分布中生成误差项；如果误差分布未知，则从残差的经验分布中重抽样。
        4.  **Wild Bootstrap**：如果时间序列存在异方差，并且其形式未知，Wild Bootstrap（有时与Sieve Bootstrap结合）可以用来生成对异方差鲁棒的Bootstrap样本。
        5.  **相序打乱（Phase Scrambling）**：一种处理时间序列的方法，通过打乱数据的傅立叶变换相位来生成新的序列，同时保持原始序列的幅度和边际分布特征。
    *   Stata等统计软件通常提供专门处理时间序列数据的Bootstrap命令（如 `tsboot` 或 `bootstrap` 结合 `cluster` 选项），可以方便地实现这些方法。

*   **面板数据（Panel Data）**：
    *   **问题**：面板数据通常包含“个体效应”（如固定效应或随机效应），导致同一组内的观测值之间存在相关性，而不同组之间通常独立 。
    *   **解决方案**：
        1.  **聚类Bootstrap（Cluster Bootstrap / Resampling by Cluster）**：这是处理面板数据最常用的方法。它不是重抽样单个观测值，而是重抽样**整个组（Cluster）**，例如重抽样个体（id）或公司 。每次抽取一个组（及其所有时间点的观测值），有放回地组合起来形成Bootstrap样本。这种方法能够保留组内的依赖性，从而得到对组内相关性鲁棒的标准误估计 。
            *   例如，在Stata中，使用 `vce(bs, cluster(id))` 选项就是实现了聚类Bootstrap 。
        2.  **残差Bootstrap（Residual Bootstrap）**：如果面板模型假设误差项在去除个体效应后是IID的，可以先估计模型（如固定效应模型FE），得到残差，然后对残差进行重抽样，再结合原始的解释变量和估计系数来生成新的因变量。但这种方法对模型假设（特别是残差IID）的依赖性较强。

在实际操作中，对于时间序列和面板数据，通常推荐首先使用**聚类Bootstrap**（针对面板数据中的组效应）或**块Bootstrap**（针对时间序列中的时间依赖性），因为它们在处理依赖性方面具有较好的鲁棒性，且对模型的具体参数形式假设较少。在特殊情况下，可以考虑Sieve Bootstrap或Wild Bootstrap等更复杂的方法。

Bootstrapping（自助法）是一种重抽样方法，它为基于渐近公式的推断提供了一种替代方案。自助样本是从我们自身的数据中抽取得到的样本。具体来说，若我们有一个样本量为 $N$ 的样本，我们会将该样本视为总体，并从中进行有放回的重复抽样。自助标准误指的是在多次此类抽样过程中，估计量的标准差。从直观上理解，我们认为通过自身数据抽样构建的抽样分布，能够很好地近似我们所追求的真实抽样分布。

对回归估计量进行自助抽样的方法有多种。最简单的一种是对 $(y_i; X_i)$ 值对进行抽样，这种方法有时被称为“配对自助法”(pairs bootstrap) 或“非参数自助法”(nonparametric bootstrap)。此外，我们也可以固定 $X_i$ 值，从残差 $(b_e_i)$ 的分布中进行抽样，并基于特定观测值的预测值和抽到的残差，构建因变量的新估计值。该方法属于“参数自助法”(parametric bootstrap) 的一种，它模拟了在非随机回归元下的抽样过程，同时确保了 $X_i$ 与回归残差之间相互独立。但另一方面，如果我们关注的是异方差性下的标准误，那么就不希望 $X_i$ 与回归残差之间存在独立性。此时，另一种残差自助法——即“野生自助法”(wild bootstrap) ——便可以发挥作用。



## 基本概念


在介绍Bootstrap之前，我们有必要回顾一些基本概念，包括“总体(population)”、“(抽样)样本(sample)”、“经验样本(empirical sample)”、“标准差(standard deviation)”和“标准误(standard error)”、“置信区间”，以及“p值”。这些基本概念对于理解Bootstrap中的一些关键技术颇有裨益。


这里我们要指出一种概念：随机样本中的每一个观察值都可被视为描述母体的一个独立信息，因此每一个观察值都带来「全新的」的信息。相对的，若样本观察值彼此不独立的话，则它们所包含的讯息就不能说是「全新的」，而是彼此有所重迭，这自然不利于达到「从样本推论母体」的目的。样本观察值彼此相关程度越高，信息重迭的越严重，越不利于统计推论。

The wild bootstrap, proposed originally by $\mathrm{Wu}(1986)$, [31] is suited when the model exhibits heteroskedasticity. The idea is, as the residual bootstrap, to leave the regressors at their sample value, but to resample the response variable based on the residuals values. That replicate, one computes a new $y$ based on
$$
y_i^*=\widehat{y}_i+\widehat{\varepsilon}_i v_i
$$
so the residuals are randomly multiplied by a random variable $v_i$ with mean 0 and variance 1 . For most distributions of $v_i$ (but not Mammen's), this method assumes that the 'true' residual distribution is symmetric and can offer advantages over simple residual sampling for smaller sample sizes. Different forms are used for the random variable $v_i$, such as
- The standard normal distribution
- A distribution suggested by Mammen (1993). [32]
$$
v_i= \begin{cases}-(\sqrt{5}-1) / 2 & \text { with probability }(\sqrt{5}+1) /(2 \sqrt{5}), \\ (\sqrt{5}+1) / 2 & \text { with probability }(\sqrt{5}-1) /(2 \sqrt{5})\end{cases}
$$
Approximately, Mammen's distribution is:
$$
v_i=\left\{\begin{array}{lll}
-0.6180 & \text { (with a } 0 \text { in the units' place) } & \text { with probability } 0.7236 \\
+1.6180 & \text { (with a } 1 \text { in the units' place) } & \text { with probability } 0.2764
\end{array}\right.
$$
- Or the simpler distribution, linked to the Rademacher distribution:
$$
v_i= \begin{cases}-1 & \text { with probability 1/2 } \\ +1 & \text { with probability 1/2 }\end{cases}
$$

In this section, we discuss the wild cluster bootstrap, which was proposed in CGM (2008) and the validity of which was proved in Djogbenou, MacKinnon, and Nielsen (2018). It is a generalization of the ordinary wild bootstrap, which was developed in Liu (1988) for the nonclustered, heteroskedastic case, following a suggestion in Wu (1986) and commentary thereon by Beran (1986). Härdle and Mammen (1993) appear to have been the first to call the method "wild", observing that it "... can be thought of as attempting to reconstruct the distribution of each residual through the use of one single observation."

The bootstrap method most deeply embedded in Stata-via the bootstrap prefix command-is a nonparametric bootstrap that, when applied to regression models, is often called the "pairs bootstrap". In the estimation context set out in the previous section, this bootstrap would typically operate by resampling $\left(\mathbf{y}_g, \mathbf{X}_g\right)$ pairs at the cluster level (Field and Welsh 2007). This is asymptotically valid, but some of the bootstrap samples may not resemble the actual sample, especially if the cluster sizes vary a lot. In a difference-in-differences context with few treated clusters, the treatment dummy could even equal zero for all observations in some bootstrap samples. For these reasons, the pairs cluster bootstrap can work poorly in some cases; see Bertrand, Duflo, and Mullainathan (2004) and MacKinnon and Webb (2017a).
