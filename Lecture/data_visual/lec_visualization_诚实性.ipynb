# 诚实性：财经媒体中的常见误导陷阱

数据可视化并不是一个“中性容器”。同一组数据，经过不同的坐标轴设定、时间窗口选择、比例映射方式和图形布局处理后，读者看到的“结论”可能完全不同。很多误导图形并没有篡改原始数据，但它们改变了数据到视觉元素之间的映射关系，从而影响了读者的判断。

从形式上看，图表就是把数值 $x$ 映射为位置、长度、面积、角度、颜色或体积等视觉对象；如果这种映射关系不再保持可比性，那么图表即使“数据真实”，也可能“表达失真”。

$$
\text{visual impression} \neq \text{data magnitude}
$$

因此，阅读财经媒体、研究报告、券商研报和上市公司公告时，不能只看“图形长什么样”，还要看“图形是怎么画出来的”。

## 1. 截断 Y 轴夸大差异

这是最常见的一类误导。比如，某公司净利润从 100 亿元增长到 103 亿元，真实增长率只有 3%。如果柱状图的纵轴从 98 而不是从 0 开始，那么柱体高度差会被明显放大，视觉上仿佛出现了“跳升”。

这一问题的核心在于：柱状图所表达的主要信息是“长度差异”，而长度的比较通常默认以零点为基准。当零点被人为抬高后，视觉比例就被扭曲了。

判断方法很简单：先看 Y 轴是否从 0 开始，然后问自己一句话：**如果纵轴从 0 开始，这个差异还会显得如此巨大吗？**

## 2. 选择性时间窗口，制造“趋势”

同一条时间序列，只要起点和终点选得不同，叙事就可能完全改变。某产品如果恰好从低点开始计算，便很容易得到“过去三年累计收益率 60%”这样的醒目结论；但如果把时间窗口向前延长几年，读者就会发现，这可能只是一次从前期大跌中的恢复，而不是持续稳定的高回报。

这一类误导并不一定来自虚构数据，而是来自**选择对自己有利的样本区间**。因此，时间序列图不能只看斜率和终点，更要看它的起点是如何被选定的。

判断方法包括两条：

* 把时间窗口适当拉长，看“漂亮结论”是否仍然成立；
* 用多个起始点重复计算累计收益率、增长率或变化率，检查结论是否稳健。

## 3. 不等距时间刻度，制造伪趋势

这类误导比截断 Y 轴更隐蔽。图上的横轴看起来像是按时间顺序均匀展开的，但实际上各区间长度并不相同，却被画成了相同宽度。这样一来，读者会自然地把最后一段的高低变化理解为与前面各段完全可比的趋势变化。

你给出的 Nobel Prize 例子正属于这一类。前面若干区间都是 10 年，而最后一个区间只有 4 年，但图上却占据了和前面相同的宽度，于是读者会误以为最近一期出现了显著下降。问题不在于最后一个数据点一定错误，而在于**它与前面的区间并不可直接比较**。

因此，时间序列图除了要求“有年份标签”之外，还要求“时间分组可比”。如果每个组的时间跨度不同，那么图上的斜率、拐点和跌幅都可能带有误导性。

## 4. 面积或体积误用

当图中用圆、球、图标大小来代表数值时，最容易出现的错误是：设计者让**半径**直接对应数值，而读者实际感受到的却是**面积**甚至**体积**。这样一来，大值会被严重夸大。

如果用圆面积表示数值 $v$，那么正确关系应满足

$$
A \propto v
$$

而圆面积满足

$$
A = \pi r^2
$$

因此半径应当满足

$$
r \propto \sqrt{v}
$$

如果错误地令 $r \propto v$，那么就会导致

$$
A \propto v^2
$$

也就是说，数值翻倍，面积会变成 4 倍；数值变为 4 倍，面积会膨胀为 16 倍。很多“气泡图很震撼”的来源，其实不是数据本身，而是比例映射出了问题。

## 5. 用双 Y 轴制造虚假相关性

双 Y 轴图形的危险之处在于：两个变量虽然单位不同，但通过分别调整左轴和右轴的上下界，可以让两条曲线看起来“高度同步”，也可以让它们看起来“几乎无关”。也就是说，视觉相关性在很大程度上取决于坐标轴设定，而不完全取决于数据本身。

设两个序列分别为 $x_t$ 和 $y_t$。图形绘制时，常常实际展示的是它们经过线性变换后的结果：

$$
x_t^\ast = a_1 + b_1 x_t,\qquad
y_t^\ast = a_2 + b_2 y_t
$$

只要适当选择 $a_1, b_1, a_2, b_2$，两条线的视觉形状就可能明显靠拢，或者被人为拉开。于是，读者很容易从“走势接近”推断出“存在强相关甚至因果关系”，这在实证分析中是非常危险的。

更稳妥的做法包括：

* 把两个序列统一转换为指数形式，例如都以起点记为 100；
* 或者分别作图；
* 或者直接报告相关系数、回归结果，而不是只依赖视觉印象。

## 6. 忽略基数，只强调百分比变化

财经媒体中经常出现“同比增长 100%”“环比翻倍”之类的说法，但如果不同时报告基数，这类表述很容易误导。因为增长率定义为

$$
g = \frac{y_1 - y_0}{y_0}\times 100%
$$

当 $y_0$ 很小时，即使绝对变化极小，也可能得到惊人的增长率。

例如，从 1 增加到 2，增长率是 100%；而从 100 增加到 101，增长率只有 1%。但前者的绝对增量只有 1，后者的绝对增量同样也是 1。只看百分比，不看基数，就会让读者把“小基数上的正常波动”误认为“重大增长”。

因此，涉及增长率、同比、环比、收益率时，最好同时呈现：

* 初始水平；
* 变化后的水平；
* 绝对增量；
* 相对增幅。

## 7. 阅读财经图表时的基本检查清单

面对一张图表，至少应依次检查以下问题：

* 坐标轴的起点是否合理，尤其是柱状图的 Y 轴是否被截断；

* 时间窗口是如何选择的，是否只截取了对结论有利的区间；

* 横轴分组是否等距，时间跨度是否可比；

* 面积、体积、气泡大小是否与数据保持正确比例；

* 双 Y 轴是否被用来强化某种“同步性”叙事；

* 百分比变化是否同时报告了对应的基数和绝对量。

很多图形的问题不在于“数据造假”，而在于“视觉编码不诚实”。对数据分析者而言，关键不只是会画图，更是能够判断一张图是否忠实地表达了数据。





# Python 示例代码

下面这份代码会一次性生成 6 张示例图，并保存到 `figs/` 文件夹中。你可以直接放到 `.ipynb` 的一个代码单元中运行，也可以保存为独立的 `.py` 文件运行。

## 1. 算法说明

1. **Step 1**: 构造 6 组小型示例数据，分别对应 6 类常见误导。
2. **Step 2**: 对每一类误导，同时绘制“相对诚实”的版本和“带有误导性”的版本。
3. **Step 3**: 将图形统一保存到 `figs/` 文件夹，便于在讲义中直接插入。
4. **Step 4**: 根据文件名，把图片插入到对应小节下方。

## 2. 完整代码



````python

# ============================================================
# 诚实性：财经媒体中的常见误导陷阱
# 作用：
#   生成 6 张示例图，分别展示常见图形误导方式
# 输出：
#   figs/fig01_truncated_y.png
#   figs/fig02_time_window.png
#   figs/fig03_irregular_scale.png
#   figs/fig04_area_misuse.png
#   figs/fig05_dual_axis.png
#   figs/fig06_percentage_base.png
# ============================================================

from pathlib import Path

import numpy as np
import matplotlib.pyplot as plt

# ------------------------------------------------------------
# 基本绘图设置
# ------------------------------------------------------------
# 说明：
# - font.sans-serif：尽量兼容 Windows 环境中的中文字体
# - axes.unicode_minus=False：避免负号显示为方块
# ------------------------------------------------------------
plt.rcParams["font.sans-serif"] = [
    "Microsoft YaHei",
    "SimHei",
    "Arial Unicode MS",
    "DejaVu Sans"
]
plt.rcParams["axes.unicode_minus"] = False
plt.rcParams["figure.dpi"] = 120

# ------------------------------------------------------------
# 创建输出目录
# ------------------------------------------------------------
OUTDIR = Path("figs")
OUTDIR.mkdir(exist_ok=True)

# ------------------------------------------------------------
# 统一的保存函数
# ------------------------------------------------------------
def save_fig(fig, filename):
    """
    保存图片，并自动调整边距。
    """
    fig.tight_layout()
    fig.savefig(OUTDIR / filename, dpi=300, bbox_inches="tight")
    plt.show()
    plt.close(fig)
````

````python


# ============================================================
# 图 1：截断 Y 轴夸大差异
# ============================================================
def plot_truncated_y():
    """
    左图：Y 轴从 0 开始，属于相对诚实的表达；
    右图：Y 轴从 99 开始，视觉上会夸大差异。
    """
    quarters = ["Q1", "Q2", "Q3", "Q4", "Q5"]
    revenue = np.array([100, 101, 102, 103, 104])

    fig, axes = plt.subplots(1, 2, figsize=(12, 4.5), sharex=True)

    # -------------------------
    # 左图：相对诚实版本
    # -------------------------
    axes[0].bar(quarters, revenue)
    axes[0].set_title("诚实版本：Y 轴从 0 开始")
    axes[0].set_ylabel("营收（亿元）")
    axes[0].set_ylim(0, 110)

    for i, v in enumerate(revenue):
        axes[0].text(i, v + 1, f"{v}", ha="center", va="bottom", fontsize=10)

    # -------------------------
    # 右图：误导版本
    # -------------------------
    axes[1].bar(quarters, revenue)
    axes[1].set_title("误导版本：Y 轴从 99 开始")
    axes[1].set_ylabel("营收（亿元）")
    axes[1].set_ylim(99, 104.5)

    for i, v in enumerate(revenue):
        axes[1].text(i, v + 0.05, f"{v}", ha="center", va="bottom", fontsize=10)

    # -------------------------
    # 图注
    # -------------------------
    fig.suptitle("截断 Y 轴会显著放大微小差异", fontsize=14)
    fig.text(
        0.5, -0.02,
        "同样是从 100 增加到 104，左图看起来变化平缓，右图却像是大幅增长。",
        ha="center", fontsize=10
    )

    save_fig(fig, "fig01_truncated_y.png")

    
````

````python


# ============================================================
# 图 2：选择性时间窗口
# ============================================================
def plot_time_window():
    """
    左图：展示完整样本期；
    右图：只从低点开始截取子样本，从而制造“高增长”叙事。
    """
    years = np.arange(2016, 2025)
    nav = np.array([118, 112, 104, 95, 80, 85, 96, 110, 128])

    # 只选取 2020 年及以后，恰好从低点附近开始
    mask = years >= 2020
    years_short = years[mask]
    nav_short = nav[mask]

    cumulative_return = (nav_short[-1] / nav_short[0] - 1) * 100

    fig, axes = plt.subplots(1, 2, figsize=(12, 4.5), sharey=True)

    # -------------------------
    # 左图：完整窗口
    # -------------------------
    axes[0].plot(years, nav, marker="o", linewidth=2)
    axes[0].set_title("完整窗口：2016 - 2024")
    axes[0].set_ylabel("净值指数")
    axes[0].set_xlabel("年份")
    axes[0].grid(alpha=0.3)

    # -------------------------
    # 右图：选择性窗口
    # -------------------------
    axes[1].plot(years_short, nav_short, marker="o", linewidth=2)
    axes[1].set_title("选择性窗口：2020 - 2024")
    axes[1].set_xlabel("年份")
    axes[1].grid(alpha=0.3)

    axes[1].annotate(
        f"累计收益率 = {cumulative_return:.1f}%",
        xy=(years_short[-1], nav_short[-1]),
        xytext=(years_short[1], nav_short[-1] - 10),
        arrowprops=dict(arrowstyle="->"),
        fontsize=10
    )

    fig.suptitle("同一产品，不同时间窗口会讲出不同故事", fontsize=14)
    fig.text(
        0.5, -0.02,
        "右图从低点附近开始，容易制造“过去几年表现非常亮眼”的印象；左图显示它更像是一次恢复。",
        ha="center", fontsize=10
    )

    save_fig(fig, "fig02_time_window.png")


# ============================================================
# 图 3：不等距时间刻度
# ============================================================
def plot_irregular_scale():
    """
    左图：最后一个区间只有 4 年，但与前面的 10 年区间占据相同宽度；
    右图：改为完整的 10 年区间后，趋势判断发生变化。
    """
    labels_left = [
        "1901-\n1910", "1911-\n1920", "1921-\n1930", "1931-\n1940",
        "1941-\n1950", "1951-\n1960", "1961-\n1970", "1971-\n1974"
    ]
    values_left = np.array([2, 4, 3, 10, 14, 29, 26, 14])

    labels_right = [
        "1901-\n1910", "1911-\n1920", "1921-\n1930", "1931-\n1940",
        "1941-\n1950", "1951-\n1960", "1961-\n1970", "1971-\n1980"
    ]
    values_right = np.array([2, 4, 3, 10, 14, 29, 26, 35])

    fig, axes = plt.subplots(1, 2, figsize=(12, 4.8), sharey=True)

    axes[0].plot(labels_left, values_left, marker="o", linewidth=2)
    axes[0].set_title("误导版本：最后一组仅 4 年")
    axes[0].set_ylabel("数量")
    axes[0].grid(axis="y", alpha=0.3)

    axes[1].plot(labels_right, values_right, marker="o", linewidth=2)
    axes[1].set_title("校正版本：使用完整 10 年区间")
    axes[1].grid(axis="y", alpha=0.3)

    fig.suptitle("不等距时间分组会制造伪下降或伪上升", fontsize=14)
    fig.text(
        0.5, -0.03,
        "左图中最后一组只有 4 年，却被画成与前面各个 10 年区间相同的宽度，因而容易造成趋势误判。",
        ha="center", fontsize=10
    )

    save_fig(fig, "fig03_irregular_scale.png")


# ============================================================
# 图 4：面积误用
# ============================================================
def plot_area_misuse():
    """
    左图：气泡面积与数值成正比；
    右图：错误地让半径与数值成正比，从而导致面积被二次放大。
    """
    labels = ["A 公司", "B 公司", "C 公司"]
    values = np.array([10, 20, 40])
    x = np.arange(len(values))
    y = np.ones_like(x)

    # -------------------------
    # 正确做法：
    # scatter 中的 s 参数本身就是面积
    # 因此只需令面积 s 与数值成正比即可
    # -------------------------
    size_honest = values * 80

    # -------------------------
    # 错误做法：
    # 如果把半径设置为与数值成正比，
    # 则面积会与数值的平方成正比
    # 在 scatter 中，面积 s 应写成“半径的平方”
    # -------------------------
    radius_wrong = values * 3.5
    size_wrong = radius_wrong ** 2

    fig, axes = plt.subplots(1, 2, figsize=(12, 4.5), sharey=True)

    axes[0].scatter(x, y, s=size_honest, alpha=0.7)
    axes[0].set_title("诚实版本：面积 ∝ 数值")
    axes[0].set_xticks(x)
    axes[0].set_xticklabels(labels)
    axes[0].set_yticks([])
    axes[0].set_xlim(-0.7, 2.7)

    for i, v in enumerate(values):
        axes[0].text(i, 1.18, f"{v}", ha="center", fontsize=11)

    axes[1].scatter(x, y, s=size_wrong, alpha=0.7)
    axes[1].set_title("误导版本：半径 ∝ 数值")
    axes[1].set_xticks(x)
    axes[1].set_xticklabels(labels)
    axes[1].set_yticks([])
    axes[1].set_xlim(-0.7, 2.7)

    for i, v in enumerate(values):
        axes[1].text(i, 1.18, f"{v}", ha="center", fontsize=11)

    fig.suptitle("错误的面积映射会严重夸大大数值", fontsize=14)
    fig.text(
        0.5, -0.02,
        "三家公司的真实数值之比为 1 : 2 : 4；若错误地让半径对应数值，视觉上的面积差异会远大于真实差异。",
        ha="center", fontsize=10
    )

    save_fig(fig, "fig04_area_misuse.png")


# ============================================================
# 图 5：双 Y 轴制造虚假相关性
# ============================================================
def plot_dual_axis():
    """
    用同一组数据，画出两张双 Y 轴图：
    - 左图：通过调整右轴范围，让两条线看起来高度同步；
    - 右图：通过放宽右轴范围，让同样的数据看起来关系很弱。
    """
    months = np.arange(1, 13)

    # 例子中的两个序列可以理解为：
    # - 左轴：某公司股价
    # - 右轴：某关键词搜索指数
    # 两者可能只是大体上同向波动，并不能据此推出因果关系
    price = np.array([100, 103, 104, 102, 105, 108, 110, 109, 111, 114, 113, 116])
    search = np.array([50, 55, 56, 53, 58, 60, 63, 61, 65, 68, 66, 70])

    fig, axes = plt.subplots(1, 2, figsize=(13, 4.8))

    # -------------------------
    # 左图：人为调节右轴，使两条线看起来很“同步”
    # -------------------------
    ax1 = axes[0]
    ax2 = ax1.twinx()

    line1 = ax1.plot(months, price, marker="o", linewidth=2, label="股价")[0]
    line2 = ax2.plot(months, search, marker="s", linewidth=2, label="搜索指数")[0]

    ax1.set_title("同一数据：看起来“高度同步”")
    ax1.set_xlabel("月份")
    ax1.set_ylabel("股价")
    ax2.set_ylabel("搜索指数")

    ax1.set_ylim(98, 118)
    ax2.set_ylim(48, 72)
    ax1.grid(alpha=0.3)

    axes[0].legend([line1, line2], ["股价", "搜索指数"], loc="upper left")

    # -------------------------
    # 右图：同样的数据，放宽右轴范围后，
    # 看起来又不那么“同步”了
    # -------------------------
    ax1 = axes[1]
    ax2 = ax1.twinx()

    line1 = ax1.plot(months, price, marker="o", linewidth=2, label="股价")[0]
    line2 = ax2.plot(months, search, marker="s", linewidth=2, label="搜索指数")[0]

    ax1.set_title("同一数据：看起来关系较弱")
    ax1.set_xlabel("月份")
    ax1.set_ylabel("股价")
    ax2.set_ylabel("搜索指数")

    ax1.set_ylim(98, 118)
    ax2.set_ylim(0, 200)
    ax1.grid(alpha=0.3)

    axes[1].legend([line1, line2], ["股价", "搜索指数"], loc="upper left")

    fig.suptitle("双 Y 轴的视觉相关性在很大程度上取决于坐标轴设定", fontsize=14)
    fig.text(
        0.5, -0.03,
        "左右两图使用的是同一组数据，但只要调整右轴范围，读者对“相关性”的直观判断就会显著变化。",
        ha="center", fontsize=10
    )

    save_fig(fig, "fig05_dual_axis.png")


# ============================================================
# 图 6：忽略基数，只强调百分比变化
# ============================================================
def plot_percentage_base():
    """
    左图：只给出增长率，容易误导；
    右图：同时展示原始水平，读者才能正确理解。
    """
    labels = ["产品 A", "产品 B"]
    before = np.array([1, 100])
    after = np.array([2, 101])
    growth_pct = (after - before) / before * 100

    fig, axes = plt.subplots(1, 2, figsize=(12, 4.8))

    # -------------------------
    # 左图：只强调增长率
    # -------------------------
    axes[0].bar(labels, growth_pct)
    axes[0].set_title("只看增长率")
    axes[0].set_ylabel("增长率（%）")
    axes[0].set_ylim(0, 110)

    for i, v in enumerate(growth_pct):
        axes[0].text(i, v + 3, f"{v:.0f}%", ha="center", fontsize=11)

    # -------------------------
    # 右图：展示基数与变化后水平
    # -------------------------
    x = np.arange(len(labels))
    width = 0.35

    axes[1].bar(x - width / 2, before, width=width, label="变化前")
    axes[1].bar(x + width / 2, after, width=width, label="变化后")

    axes[1].set_title("同时看基数与变化后水平")
    axes[1].set_ylabel("数量")
    axes[1].set_xticks(x)
    axes[1].set_xticklabels(labels)
    axes[1].legend()

    for i, (b, a) in enumerate(zip(before, after)):
        axes[1].text(i - width / 2, b + 2, f"{b}", ha="center", fontsize=10)
        axes[1].text(i + width / 2, a + 2, f"{a}", ha="center", fontsize=10)

    fig.suptitle("百分比变化必须结合基数一起阅读", fontsize=14)
    fig.text(
        0.5, -0.02,
        "产品 A 从 1 增至 2，增长率为 100%；产品 B 从 100 增至 101，增长率只有 1%。若只看百分比，很容易误判实际重要性。",
        ha="center", fontsize=10
    )

    save_fig(fig, "fig06_percentage_base.png")


# ============================================================
# 主程序：依次生成全部图片
# ============================================================
if __name__ == "__main__":
    plot_truncated_y()
    plot_time_window()
    plot_irregular_scale()
    plot_area_misuse()
    plot_dual_axis()
    plot_percentage_base()

    print("全部图片已生成，保存在 figs/ 文件夹中。")

````
